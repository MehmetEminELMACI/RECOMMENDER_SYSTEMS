# <font color='red'> It predicts the scores of the movies that the user did not give according to the movies that the user gave their ratings.

```python
import pandas as pd
```

```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3))
print (ratings.head())
m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, encoding="iso-8859-1", usecols=range(2))
print (movies.head())
```

```python
ratings = pd.merge(movies, ratings)
```

```python
ratings.head()
```

```python
ratings.title.nunique(), ratings.user_id.nunique()
```

```python
movieRatings = ratings.pivot_table(index=["user_id"], columns=["title"], values="rating")
movieRatings.head()
# Ratings given to movies by each user
```

```python
starWarsRatings = movieRatings["Star Wars (1977)"]
starWarsRatings.head()
# Star wars movie ratings
```

```python
starWarsRatings.value_counts(dropna=False)
```

```python
movieRatings[['101 Dalmatians (1996)', 'Star Wars (1977)']].corr() # İki film arasındaki benzerlik
```
